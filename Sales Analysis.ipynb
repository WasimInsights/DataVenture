## Creating group 
### df.groupby(['Gender'], as_index=False): This groups the DataFrame df by the ‘Gender’ column. The as_index=False keeps the ‘Gender’ as a column, not an index.
### ['Amount'].sum(): This calculates the sum of the ‘Amount’ for each group (i.e., ‘Gender’).
### sort_values(by='Amount', ascending=False): This sorts the resulting DataFrame by the ‘Amount’ column in descending order (highest to lowest).
### The entire line of code is used to find the total ‘Amount’ for each ‘Gender’ and then sort these totals in descending order.
#### refer code below

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x=  'Gender', y = 'Amount', data = sales_gen)
```

#### **From the above output we can conclude that the most of the buyers are females and purchasing power of females is graeter than males

# Total Amount vs Age

```python
sales_age = df.groupby(['Age Group', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Gender')
```

###  The majority of the buyers belong to the age group of 25 to 35 years. This suggests that our products or services are particularly appealing to this demographic.
### Within this age group, females constitute a higher proportion of the buyers. This indicates a strong female customer base in this age segment.
#### These insights are crucial as they can guide our marketing and product development strategies. This could potentially lead to increased customer engagement and sales. It’s important to note that while these trends provide valuable insights, customer preferences can change over time, and it’s essential to continually analyze customer data to stay updated.

```python
df.columns
```

##  Top 10  staes by orders
## sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10): This line of code does several things
### 1)Groups the data by ‘State’.
### 2) Sums up the ‘Orders’ for each ‘State’.
### 3)Sorts the states by the total ‘Orders’ in descending order.
### 4)Selects the top 10 states with the most orders.
## sns.set(rc={'figure.figsize':(width,height)}): 
### This sets the size of the figure for the plot
## sns.barplot(x='State', y='Orders', data=sales_state): 
### This creates a bar plot with ‘State’ on the x-axis and ‘Orders’ on the y-axis
###  The data for the plot comes from the sales_state DataFrame


```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data=sales_state)
```

### Top 10 states by Amount (same as above just replaced **'Orders'** with **'Amount'**

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Amount', data=sales_state)
```

### Uttar Pradesh, Maharashtra, and Karnataka lead in orders and total amount.
### Despite fewer orders, Haryana contributes more to the total amount than Kerala.These insights are crucial as they can guide our marketing and product development strategies. For instance, we can tailor our marketing campaigns to resonate more with customers in the top-performing states and consider their preferences during product development. Additionally, we can investigate why some states, despite having fewer orders, contribute more to the total amount and leverage this information to increase our revenue
